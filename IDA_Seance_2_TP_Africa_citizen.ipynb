Le Taux Brut de Natalité (CBR) est facile à comprendre, mais il inclut à la fois les hommes et les femmes de tous âges dans le dénominateur. Nous calculons ensuite le Taux de Fécondité Total (TFR). Contrairement au CBR, le TFR ajuste les compositions d'âge dans la population féminine. Pour ce faire, nous devons d'abord calculer le Taux de Fécondité Spécifique par Âge (ASFR), qui représente le taux de fécondité pour les femmes de la tranche d'âge reproductive [15,50). L'ASFR pour la plage d'âge [x,x+ δ), où x est l'âge de début et δ est la largeur de la plage d'âge (mesurée en années), est défini comme suit:

$ \text{ASFR}_{[x, x + \delta)} = \frac{\text{nombre de naissances de femmes d'âge } [x, x + \delta)}{\text{Nombre de personnes-années vécues par les femmes d'âge } [x, x + \delta)}$

Notez que les crochets, [ et ], incluent la limite, tandis que les parenthèses, ( et ), l'excluent. Par exemple, [20,25) représente la plage d'âge qui est supérieure ou égale à 20 ans et inférieure à 25 ans. Dans les données démographiques typiques, la plage d'âge δ est fixée à 5 ans.

Calculez l'ASFR pour la Suède et le Kenya ainsi que pour le monde entier pour chacune des deux périodes. Stockez les ASFR résultants séparément pour chaque région.

Bien que cela soit simple et facile à comprendre, le TBM ne prend pas en compte la composition par âge d'une population. Nous calculons donc le Taux de Mortalité Spécifique par Âge (TMSA).
